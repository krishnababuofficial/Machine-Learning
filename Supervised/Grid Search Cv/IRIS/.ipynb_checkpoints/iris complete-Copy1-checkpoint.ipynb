# EDA(Exploratory Data Analysis)

```python
sns.pairplot(data=ir,hue='Species')
plt.show();
```

```python
plt.figure(figsize=(13,7))
sns.boxplot(data=ir)
plt.grid()
plt.show();
```

```python
sns.boxplot(ir["sepal width in cm"])
plt.show();

q3=ir["sepal width in cm"].quantile(0.75)
q1=ir["sepal width in cm"].quantile(0.25)
iqr=q3-q1
print(f"IQR: {iqr}\n")

upper_whisker=q3+1.5*iqr
lower_whisker=q1-1.5*iqr
print(f"UPPER WHISKER: {upper_whisker}\nLOWER WHISKER: {lower_whisker}\n\n")

upper_whisker_index=ir[ir["sepal width in cm"]>upper_whisker].index
lower_whisker_index=ir[ir["sepal width in cm"]<lower_whisker].index
print(f"Upper Whisker Index :{upper_whisker_index}\nLower Whisker Index: {lower_whisker_index}\n\n")

ir.loc[upper_whisker_index,"sepal width in cm"]=np.nan
ir.fillna(upper_whisker,inplace=True)
ir.loc[lower_whisker_index,"sepal width in cm"]=np.nan
ir.fillna(lower_whisker,inplace=True)

sns.boxplot(ir["sepal width in cm"])
plt.show();
```

# Encoding

```python
le=LabelEncoder()
ir['Species']=le.fit_transform(ir['Species'])
```

```python
ir['Species'].unique()
```

### _To transform "categorical value to numerical value" encoding is done, So that it could be easily fitted to a machine learning model._

- _[0] --> Iris-versicolor_

- _[1] --> Iris-virginica_

- _[2] --> Iris-setosa_

# Model Training

```python
x=ir.iloc[:,:-1]
print(x)
y=ir.iloc[:,-1]
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
def model(mymod):
    mymod.fit(xtrain,ytrain)
    ypred=mymod.predict(xtest)
    
    train=mymod.score(xtrain,ytrain)
    test=mymod.score(xtest,ytest)
    
    print(f"Training Accuracy: {train}\nTesting Accuracy: {test}\n\n")
    print(classification_report(ytest,ypred))
    return mymod
```

```python
Knn=model(KNeighborsClassifier())
```

```python
Lr = model(LogisticRegression())
```
