```python
lp["Dependents"].replace(np.nan,0,inplace=True)
```

```python
lp["Loan_Amount_Term"].mean()
```

```python
print(lp.isna().sum())
lp
```

# 5.Skew Handeling

```python
columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term', 'Credit_History', 'Property_Area']
```

```python
for i in columns:
    print(i)
    print(skew(lp[i]))
    plt.figure()
    sns.distplot(lp[i])
    plt.show();
```

```python
lp["Dependents"]=np.sqrt(lp["Dependents"])
print(skew(lp["Dependents"]))
print(sns.distplot(lp["Dependents"]))
plt.show();
```

```python
lp["ApplicantIncome"]=np.log(lp["ApplicantIncome"])
print(skew(lp["ApplicantIncome"]))
print(sns.distplot(lp["ApplicantIncome"]))
plt.show();
```

```python
lp["LoanAmount"]=np.log(lp["LoanAmount"])
print(skew(lp["LoanAmount"]))
print(sns.distplot(lp["LoanAmount"]))
plt.show();
```

# 6. Model Training

### Splitting dataset into features and Target
- X [FEATURES]
- Y[TARGET]

```python
x=lp.iloc[:,:-1]
print(x)
y=lp.iloc[:,-1]
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    
    return model
```

```python
#make variabel for save the result and to show it

classifiers=['Logistic Regression','Random Forest','SVM','Decission Tree','K-Nearest Neighbor']
y_position=np.arange(len(classifiers))
score=[]
```

###  _Logistic Regression_

```python
lr=mymodel(LogisticRegression())
```

```python
scores=cross_val_score(lr,xtest,ytest,cv=5)
score.append(scores.mean())
print('The Accuracy of Logistic Regression is %.2f%%' %(scores.mean()*100))
```

### _Random Forest_

```python
rf=mymodel(RandomForestClassifier(n_estimators=1000)) 
```

```python
scores=cross_val_score(rf,xtest,ytest,cv=5)
score.append(scores.mean())
print('The Accuracy of Random Forest classification is %.2f%%' %(scores.mean()*100))
```

### _SVM_

```python
svm=mymodel(SVC(kernel='linear'))    
```

```python
scores=cross_val_score(svm,xtest,ytest,cv=5)
score.append(scores.mean())
print('The Accuracy of svm classification is %.2f%%' %(scores.mean()*100))
```

### _Decission Tree_

```python
dt=mymodel(DecisionTreeClassifier())
```

```python
scores=cross_val_score(dt,xtest,ytest,cv=5)
score.append(scores.mean())
print('The Accuracy of Decission Tree classification is %.2f%%' %(scores.mean()*100))
```

### _KNeighbors(Knn)_

```python
knn=mymodel(KNeighborsClassifier())
```

```python
scores=cross_val_score(knn,xtest,ytest,cv=5)
score.append(scores.mean())
print('The Accuracy for knn classification is %.2f%%' %(scores.mean()*100)) 
```

# 7. Result 

```python
plt.barh(y_position,score,align='center',alpha=0.4)
plt.yticks(y_position,classifiers)
plt.xlabel('Score')
plt.title('Classificaton Performance')
plt.show();
```

#### The Logistic Regression and Random forest are having the highest score from other Classification Algorithm
